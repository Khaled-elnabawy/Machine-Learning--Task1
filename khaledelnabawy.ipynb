```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python

data = pd.read_csv("/content/salaries(1).csv")
data
```

```python
inputs = data.drop('has_diabetes', axis='columns')
target = data['has_diabetes']
inputs
```

```python
le_age = LabelEncoder()
le_bmi = LabelEncoder()
le_family = LabelEncoder()
le_glucose = LabelEncoder()

inputs['age_group_n'] = le_age.fit_transform(inputs['age_group'])
inputs['bmi_category_n'] = le_bmi.fit_transform(inputs['bmi_category'])
inputs['family_history_n'] = le_family.fit_transform(inputs['family_history'])
inputs['glucose_level_n'] = le_glucose.fit_transform(inputs['glucose_level'])

inputs_n = inputs.drop(['age_group', 'bmi_category', 'family_history', 'glucose_level'], axis='columns')
inputs_n
```

```python
model = tree.DecisionTreeClassifier()
model.fit(inputs_n, target)
```

```python
import pandas as pd

input_data = [[2, 1, 0, 1]]
input_df = pd.DataFrame(input_data, columns=['age_group_n', 'bmi_category_n', 'family_history_n', 'glucose_level_n'])
prediction = model.predict(input_df)
print("Prediction 1:", prediction)

input_data = [[0, 0, 1, 0]]
input_df = pd.DataFrame(input_data, columns=['age_group_n', 'bmi_category_n', 'family_history_n', 'glucose_level_n'])
prediction = model.predict(input_df)
print("Prediction 2:", prediction)
```
